```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
import sympy as sp

x = sp.symbols('x')
equation = sp.Eq(x**2, 1)
solutions = sp.solve(equation, x)
solutions
```
